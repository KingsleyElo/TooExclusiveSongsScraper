```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files

def scrape_too_exclusive(base_url, total_pages):
    music_details_link = []

    for page_number in range(1, total_pages + 1):
        page_url = f"{base_url}/page/{page_number}"
        response = requests.get(page_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        head = soup.find('div', class_="jeg_posts jeg_load_more_flag").find_all('article', class_="jeg_post jeg_pl_lg_2 format-standard")
        for link in head:
            href = link.find('a').get('href')
            music_details_link.append(href)
            print(href)

    music_song_description = []
    music_song_image = []
    music_song_download_link = []
    song_artist_name = []

    for item in music_details_link:
        response1 = requests.get(item)
        soup = BeautifulSoup(response1.content, 'html.parser')

        # Song Description
        new = soup.find('div', class_="entry-content no-share").find('div', class_="content-inner").find_all('p')
        text_content = [paragraph.get_text() for paragraph in new]
        full_text = ' '.join(text_content)
        music_song_description.append(full_text)

        # Song Image
        new = soup.find('div', class_="entry-content no-share")
        if new:
            new_inner = new.find('div', class_="content-inner")
            if new_inner:
                img_tag = new_inner.find('img')
                if img_tag and 'data-src' in img_tag.attrs:
                    song_image = img_tag['data-src']
                    music_song_image.append(song_image)
                else:
                    music_song_image.append('No image found')
            else:
                music_song_image.append('No image found')
        else:
            music_song_image.append('No image found')

        # Song Download Link
        new = soup.find('div', class_="entry-content no-share").find('div', class_="content-inner")
        if new:
            download_link_tag = new.find('a', href=True)
            if download_link_tag and 'href' in download_link_tag.attrs:
                download_link = download_link_tag['href']
            else:
                download_link = 'No_download_link_found'
        else:
            download_link = 'No_download_link_found'
        music_song_download_link.append(download_link)

        # Song Artist Name
        song_name = soup.find('div', class_="entry-header").find('h1').get_text()
        song_artist_name.append(song_name)

    # Create a dictionary to store the data
    data1 = {
        'Song_Image_Link': music_song_image,
        'Song_Name': song_artist_name,
        'Song_Description': music_song_description,
        'Song_Download_Link': music_song_download_link
    }

    # Convert the dictionary to a pandas DataFrame
    df = pd.DataFrame(data1)

    # Save DataFrame to an Excel file
    excel_file_path = 'Too_exclusive_songs_scraped.xlsx'
    df.to_excel(excel_file_path, index=False)

    # Download the Excel file
    files.download(excel_file_path)

    # Print the resulting DataFrame
    df.head(20)

# Example usage
base_url = "https://tooxclusive.com/main/download-mp3/"
# The number of pages scraped is dependent on you, i choose to scrape 2 pages
total_pages = 2
scrape_too_exclusive(base_url, total_pages)
```
